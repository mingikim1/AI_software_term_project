```python
model_id = "minggi777/AI_software_fine_tuning"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id, device_map="auto")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import FAISS
from langchain.document_loaders import PyPDFLoader
from langchain.schema.runnable import RunnablePassthrough
```

```python
from langchain.llms import HuggingFacePipeline
from langchain.prompts import PromptTemplate
from transformers import pipeline
from langchain.chains import LLMChain

text_generation_pipeline = pipeline(
    model=model,
    tokenizer=tokenizer,
    task="text-generation",
    temperature=0.2,
    return_full_text=True,
    max_new_tokens=128,
)

prompt_template = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.
### Instruction:
Use your knowledge of artificial intelligence to provide an appropriate response.

### Context:
{context}

### input:
{input}

### response:
 """

llm = HuggingFacePipeline(pipeline=text_generation_pipeline)

# Create prompt from prompt template
prompt = PromptTemplate(
    input_variables=["context", "input"],
    template=prompt_template,
)

# Create llm chain
llm_chain = prompt | llm
```

```python
loader = []

for i in range(1, 52):
  loader.append(PyPDFLoader("/content/drive/MyDrive/hugman_2024_ML_book-" + str(i) + ".pdf"))

doc = [] # document 별 chunk로 분할
for i in range(len(loader)):
  current_doc = loader[i].load()
  text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
  chunks = text_splitter.split_documents(current_doc)
  doc.append(chunks)
```

```python
# 모든 문서의 청크를 하나의 리스트로 합치기
all_chunks = []
for document_chunks in doc:
    all_chunks.extend(document_chunks)
```

```python
# huggingface에 있는 embedding을 가져다 씀
from langchain.embeddings import HuggingFaceEmbeddings

model_name = "sentence-transformers/LaBSE"
encode_kwargs = {'normalize_embeddings': True}
hf = HuggingFaceEmbeddings(
                            model_name=model_name,
                            encode_kwargs=encode_kwargs
                            )



db = FAISS.from_documents(all_chunks, hf) # text들을 embedding 해서 DB에 넣어줌
retriever = db.as_retriever(
                            search_types='similarity', # dot product
                            search_kwargs={'k': 4}
                            )
```

```python
from langchain_core.output_parsers import StrOutputParser

# rag_chain 정의 (retriever가 DB에서 문서를 찾아줌)
rag_chain = (
              {"context": retriever, "input": RunnablePassthrough()}
              | llm_chain
            )
```

```python
result = rag_chain.invoke("What is multi-head attention. And explain advantage and disadvantage of multihead-attention.")
```
